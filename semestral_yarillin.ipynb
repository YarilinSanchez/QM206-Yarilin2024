Desarrolle un Programa en Python para simular la cinética:

A→X+B

X→C


```python
import numpy as np
import matplotlib.pyplot as plt
```

A→X+B

X→C


```python
A = 3.0
B = 0.5
X = 0.2
C = 0.8
k1 = 0.1
k2 = 0.01
t = 0
dt = 0.1
cA = []
cB = []
cX = []
cC = []
tt = []
while t<500:
  cA.append(A)
  cB.append(B)
  cX.append(X)
  cC.append(C)
  tt.append(t)
  dA = -k1*A*A*dt
  dB = k1*A*A*dt-k2*X*dt
  dX = k1*A*A*dt-k2*X*dt
  dC = k2*X*dt
  A = A+dA
  B = B+dB
  X = X+dX
  C = C+dC
  t = t+dt
plt.plot(tt, cA, label="[A]", color='blue')
plt.plot(tt, cB, label="[B]", color='red')
plt.plot(tt, cX, label="[X]", color='green')
plt.plot(tt, cC, label="[C]", color='purple')
plt.xlabel('Tiempo')
plt.ylabel('Concentraciones')
plt.legend()
plt.title('Simulación de Cinética Química')
plt.grid()
plt.show()
```
