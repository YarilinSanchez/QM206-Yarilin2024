2. Desarrolle un Programa en Python para simular la cinética:

A→X+B

X→C


A→X+B

X→C


```python
A = 3.0
B = 0.5
X = 0.2
C = 0.8
k1 = 0.1
k2 = 0.01
t = 0
dt = 0.1
cA = []
cB = []
cX = []
cC = []
tt = []
while t<500:
  cA.append(A)
  cB.append(B)
  cX.append(X)
  cC.append(C)
  tt.append(t)
  dA = -k1*A*A*dt
  dB = k1*A*A*dt-k2*X*dt
  dX = k1*A*A*dt-k2*X*dt
  dC = k2*X*dt
  A = A+dA
  B = B+dB
  X = X+dX
  C = C+dC
  t = t+dt
plt.plot(tt, cA, label="[A]", color='blue')
plt.plot(tt, cB, label="[B]", color='red')
plt.plot(tt, cX, label="[X]", color='green')
plt.plot(tt, cC, label="[C]", color='purple')
plt.xlabel('Tiempo')
plt.ylabel('Concentraciones')
plt.legend()
plt.title('Simulación de Cinética Química')
plt.grid()
plt.show()
```

PROBLEMA 5:

Diseñar un programa en Python que analice los resultados de un laboratorio clínico y clasifique cada parámetro como bajo, normal y alto para 5 pacientes diferentes. Nota: debe investigar cuales son los parámetros analizados en un uroanálisis y los rangos correspondientes a cada parámetro.


```python
import pandas as pd
import numpy as np
```

```python
rango_valores ={"pH": (4.5, 8.0), "Densidad": (1.005, 1.030), "Proteínas (mg/dL)": (0, 20), "Glucosa (mg/dL)": (0, 0.5), "Cuerpos Cetónicos": ["Negativo"], "Bilirrubina (mg/dL)": [0], "Urobilinógeno (mg/dL)": (0.1, 1.0), "Nitritos": ["Negativo"], "Leucocitos (leu/µL)": (0, 5), "Eritrocitos (leu/µL)": (0, 2)}
def clasificar_parametro(nombre_parametro, valor):
    rango = rango_valores[nombre_parametro]
    if isinstance(rango, tuple):
        if valor < rango[0]:
            return "Bajo"
        elif rango[0] <= valor <= rango[1]:
            return "Normal"
        else:
            return "Alto"
    else:
        if valor in rango:
            return "Normal"
        else:
            return "Anormal"
```

ejemplo:

```python
data={"Paciente": [1, 2, 3, 4, 5],
    "pH": [7.0, 5.5, 8.2, 6.0, 4.0],
    "Densidad": [1.015, 1.005, 1.045, 1.025, 1.010],
    "Proteínas (mg/dL)": [10, 30, 0, 5, 50],
    "Glucosa (mg/dL)": [0, 0.1, 0.6, 0.2, 0],
    "Cuerpos Cetónicos": ["Negativo", "Positivo", "Negativo", "Negativo", "Traces"],
    "Bilirrubina (mg/dL)": [0, 0.1, 0, 0, 0],
    "Urobilinógeno (mg/dL)": [0.5, 1.5, 0.1, 0.4, 1.1],
    "Nitritos": ["Negativo", "Positivo", "Negativo", "Negativo", "Positivo"],
    "Leucocitos (leu/µL)": [4, 10, 0, 1, 6],
    "Eritrocitos (leu/µL)": [1, 3, 0, 2, 4]}
df=pd.DataFrame(data)
df["pH Clasificación"] = np.where(df['pH'] < 4.5, 'Bajo',
                                  np.where(df['pH'] <= 8.0, 'Normal', 'Alto'))

df['Densidad Clasificación'] = np.where(df['Densidad'] < 1.003, 'Bajo',
                                        np.where(df['Densidad'] <= 1.030, 'Normal', 'Alto'))

df['Proteínas Clasificación'] = np.where(df['Proteínas (mg/dL)'] < 0, 'Bajo',
                                         np.where(df['Proteínas (mg/dL)'] <= 20, 'Normal', 'Alto'))

df['Glucosa Clasificación'] = np.where(df['Glucosa (mg/dL)'] < 0, 'Bajo',
                                       np.where(df['Glucosa (mg/dL)'] <= 0.5, 'Normal', 'Alto'))

df['Cuerpos Cetónicos Clasificación'] = np.where(df['Cuerpos Cetónicos'] == 'Negativo', 'Normal', 'Anormal')

df['Bilirrubina Clasificación'] = np.where(df['Bilirrubina (mg/dL)'] == 0, 'Normal', 'Anormal')

df['Urobilinógeno Clasificación'] = np.where(df['Urobilinógeno (mg/dL)'] < 0.1, 'Bajo',
                                             np.where(df['Urobilinógeno (mg/dL)'] <= 1.0, 'Normal', 'Alto'))

df['Nitritos Clasificación'] = np.where(df['Nitritos'] == 'Negativo', 'Normal', 'Anormal')

df['Leucocitos Clasificación'] = np.where(df['Leucocitos (leu/µL)'] < 0, 'Bajo',
                                          np.where(df['Leucocitos (leu/µL)'] <= 5, 'Normal', 'Alto'))

df['Eritrocitos Clasificación'] = np.where(df['Eritrocitos (leu/µL)'] < 0, 'Bajo',
                                           np.where(df['Eritrocitos (leu/µL)'] <= 2, 'Normal', 'Alto'))
print(df)
```
